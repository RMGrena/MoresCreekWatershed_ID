```python
def Stream_Sites (row):
   if row['Type'] == 'MC' :
      return 'MC'
   if row['Type'] == 'BC' :
      return 'BC'
   if row['Site_no'] == 'C1E' :
      return 'C1E'
   if row['Site_no'] == 'LR' :
      return 'LR'
   if row['Site_no'] == 'MR' :
      return 'MR'
   if row['Site_no'] == 'UR' :
      return 'UR'
   if row['Site_no'] == 'C1W' :
      return 'C1W'
   if row['Site_no'] == 'C2E' :
      return 'C2E'
   if row['Site_no'] == 'C2M' :
      return 'C2M'
   if row['Site_no'] == 'LG' :
      return 'LG'
    
s3['Stream_Sites'] = s3.apply (lambda row: Stream_Sites(row), axis=1)
```
